# Net Promoter Score (NPS) Analysis Using NumPy

<b> Net Promoter Score (NPS) <b> is a customer loyalty metric that measures the likelihood of customers recommending a company, ranging from -100 to 100. It is calculated by subtracting the percentage of detractors (0–6 score) from promoters (9-10 score). [Link](https://www.medallia.com/net-promoter-score/)

### Formula: 
%Promoters - %Detractors


### Categorization:
- Promoters (9-10): Loyal enthusiasts who will keep buying and refer others. <br>
- Passives (7-8): Satisfied but unenthusiastic customers vulnerable to competitors. <br>
- Detractors (0-6): Unhappy customers who can damage the brand.

### Significance of NPS
0-30: Good. <BR> 30-70: Great. <br >70+: Excellent. <br> Global Average: Around +32 (based on 2025 benchmarks, [link](https://www.surveymonkey.com/curiosity/what-is-a-good-net-promoter-score/)). <br> 

### Project Overview

This project analyzes customer survey responses and calculates the Net Promoter Score (NPS) using NumPy.

The notebook:

- Segments customers into Promoters, Passives, and Detractors
- Calculates percentage shares
- Computes the final NPS score
- Validates calculations
- Vizualizes results
- Derives inferences

Skills demonstrated:
- NumPy arrays
- Boolean masking
- Fancy indexing
- Data validation

```python
#Importing numpy library/package

import numpy as np
```

```python
#Loading the text file as a numpy array with data type as integer

score=np.loadtxt("survey.txt", dtype="int")
```

```python
score
```

```python
# Calculating the length of array / number of feedbacks recieved

total_scores = len(score)
total_scores
```

```python
#Calculating the number of promoters using masking or boolean indexing

promoters=score[score>=9]
promoters
```

```python
number_of_promoters=len(promoters)
number_of_promoters
```

```python
#Calculating the number of passives using np.where which uses fancy indexing

passives = score[np.where((score==7) | (score==8))]
passives
```

```python
number_of_passives = len(passives)
number_of_passives
```

```python
#Calculating the number of detractors using masking / boolean indexing

detractors=score[score<=6]
detractors
```

```python
number_of_detractors=len(detractors)
number_of_detractors
```

```python
#Validation

number_of_promoters+number_of_passives+number_of_detractors==total_scores

#Hence, the above calculations are correct
print("Validation Successful")
```

```python
# Determining percentage share of each category

# 1) Promoters
promoters_share=(number_of_promoters/total_scores)*100
passives_share=(number_of_passives/total_scores)*100
detractors_share=(number_of_detractors/total_scores)*100
```

```python
print(f"Promoters Share: {round(promoters_share,2)}% \nPassives Share: {round(passives_share,2)}% \nDetractors Share: {round(detractors_share,2)}%")
```

```python
#Validation

promoters_share+passives_share+detractors_share==100

#Hence, the above calculations are correct
print("Validation Successful")
```

```python
#Calculating the NPS

NPS=round((promoters_share-detractors_share),2)
NPS
```

```python
print(f"The Net Promoter Score for the company is {NPS}%.") 
```

```python
#Conclusion

if NPS > 50:
    category = "Excellent"
elif NPS > 0:
    category = "Good"
else:
    category = "Needs Improvement"

print(f"NPS Score: {NPS}%")
print(f"Category: {category}")
```

```python
# Vizualization

import matplotlib.pyplot as plt

categories = ["Promoters", "Passives", "Detractors"]
shares = [promoters_share, passives_share, detractors_share]

plt.bar(categories, shares)
plt.ylabel("Percentage Share")
plt.xlabel("Categories")
plt.title("Customer Segmentation by NPS")
plt.show()
```

```python
# Function creation for NPS calculation

def calculate_NPS(a):
    promoters=len(a[a>=9])
    detractors=len(a[a<=6])
    total=len(a)
    NPS=((promoters/total)*100)-((detractors/total)*100)
    return round(NPS,2)

calculate_NPS(score)
```

- Detractor percentage is relatively low
